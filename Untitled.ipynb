```python
employee_name = "Your Name"
salary = 3800
years_of_experience = 1

if salary >=4000 and years_of_experience >= 2:
    level = "Mid Level"
elif salary >=3000:
    level = "Junior Level"
else:
    level = "Entry Level"

print(f"Employee: {employee_name}")
print(f"Level: {level}")
```

```python
employee_names = ["Hana", "Anton", "Loredana"]
monthly_salaries = [3200, 4100, 2900]

for index in range(len(employee_names)):
    name = employee_names[index]
    salary = monthly_salaries[index]

    if salary >= 3500:
        status = "Above average"
    else:
        status = "Below average"

print(f"Employee: {name}, Salary: {salary}, Status: {status}")
```
