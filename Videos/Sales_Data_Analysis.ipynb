```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("sales.csv")
print(df.head())
print(df.info())
```

```python
print(df.isnull().sum())
```

```python
df = df.dropna()
```

```python
df = df.drop_duplicates()
```

```python
print(df.describe())
```

```python
print(df['Product'].value_counts())
```

```python
total_revenue = df['Revenue'].sum()
print("Total.´l Revenue:", total_revenue)
```

```python
avg_revenue = df.groupby('Product')['Revenue'].mean()
print(avg_revenue)
```

```python
best_seller = df.groupby('Product')['Quantity'].sum().idxmax()
print("Best selling product:", best_seller)
```

```python
# Revenue per Product
sns.barplot(x='Product', y='Revenue', data=df)
plt.title("Revenue per Product")
plt.show()
```

```python
# Quantity distribution
sns.histplot(df['Quantity'], bins=5)
plt.title("Quantity Distribution")
plt.show()
```
